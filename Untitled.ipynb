```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.optimizers import Adam, SGD
import matplotlib.pyplot as plt
from classification_models.tfkeras import Classifiers
import numpy as np
import cv2
import os
```

```python
trdata = ImageDataGenerator()
```

```python
traindata = trdata.flow_from_directory(directory="Sign-Language-Digits-Dataset-master/HandDetect/train",target_size=(224,224), shuffle=True)
```

```python
tsdata = ImageDataGenerator()
```

```python
testdata = tsdata.flow_from_directory(directory="Sign-Language-Digits-Dataset-master/HandDetect/test",target_size=(224,224), shuffle=True)
```

```python
ResNet18, preprocess_input = Classifiers.get('resnet18')
base_model = ResNet18(input_shape=(224,224,3), weights='imagenet', include_top=False)
```

```python
base_model.summary()
```

```python
for layers in (base_model.layers)[:]:
    layers.trainable = False
    print(layers)
```

```python
x = tf.keras.layers.GlobalAveragePooling2D()(base_model.output)
x = Dense(1024)(x)
x = Dense(256)(x)
output = Dense(2, activation='softmax')(x)
model = Model(base_model.input, output)
```

```python
model.summary()
```

```python
model.compile(optimizer = SGD(lr=0.0001, momentum=0.9), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
checkpoint = ModelCheckpoint("Resnet18_1.h5", monitor='val_accuracy', verbose=1, save_best_only=True, save_weights_only=False, mode='auto', save_freq='epoch')
early = EarlyStopping(monitor='val_accuracy', min_delta=0, patience=30, verbose=1, mode='auto')
hist = model.fit_generator(generator= traindata, steps_per_epoch= 5, epochs= 100, validation_data= testdata, validation_steps=1, callbacks=[checkpoint,early])
model.save_weights("Resnet18_1.h5.h5")
```

```python
plt.plot(hist.history["accuracy"])
plt.plot(hist.history['loss'])
plt.plot(hist.history["val_accuracy"])
plt.plot(hist.history['val_loss'])
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Accuracy","loss","Val_acc","Val_loss"])
plt.show()
```
